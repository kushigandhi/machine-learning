## scaling 

```python
df
```

```python
df6=df.copy()
df6
```

```python
df6=df6.fillna(method='ffill')
df6
```

```python
df6.head()
```

```python
df6['age_scaled']=df6['Age']/df6['Age'].max()
df6
```

```python
df6['Fare_scaled']=df6['Fare']/df6['Fare'].max()
df6
```

```python
from sklearn.preprocessing import MaxAbsScaler
Mas=MaxAbsScaler()
Mas
```

```python
df6[['Age','Fare']]=Mas.fit_transform(df[['Age','Fare']])
df6
```

# min max scaling

```python
df6['Age'].agg(['min','max'])
```

```python
df6['Age_scale']=(df6['Age']-df6['Age'].min())/(df6['Age'].max()-df6['Age'].min())
df6
```

```python
df6['Fare_scaled']=(df6['Fare']-df6['Fare'].min())/(df6['Fare'].max()-df6['Fare'].min())
df6
```

```python
from sklearn.preprocessing import MinMaxScaler
Mss=MinMaxScaler()
Mss
```

```python
df6[['Age','Fare']]=Mss.fit_transform(df[['Age','Fare']])
df6
```

## standard scaling

```python
df6['Age_scaled']=(df6['Age']-df6['Age'].mean())/(df6['Age'].std())
df6
```

```python
df6['Fare_scaled']=(df6['Fare']-df6['Fare'].mean())/(df6['Fare'].std())
df6
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
ss
```

```python
df6[['Age','Fare']]=ss.fit_transform(df[['Age','Fare']])
df6
```

# robust

```python
df6['age_sc']=(df6['Age']-df6['Age'].quantile(.25))/(df6['Age'].quantile(.75)-df6['Age'].quantile(.25))
df6
```

```python
df6['Fare_sc']=(df6['Fare']-df6['Age'].quantile(.25))/(df6['Age'].quantile(.75)-df6['Age'].quantile(.25))
df6
```

```python
from sklearn.preprocessing import RobustScaler
rs=RobustScaler()
rs
```

```python
df6[['Age','Fare']]=rs.fit_transform(df[['Age','Fare']])
df6
```

## automating all the above steps performed using pipeline() form sklearn library
### steps to create pipeline
#### 1.import all the libraries and functions need to performing data preprocessing 
#### 2.mention the steps need to perform automation and store all the steps in the 'steps' variable
#### 3.import pipeline() form sklearn library and create pipeline with 'steps' variable

```python
num_steps=[('num missing treatment',si),('num_scaling',mm)]
num_steps
```

## create column transforma() to automate pipelines 
### steps 
#### 1.seperate all numbericals nd variables
#### 2.create a column transforma() using sklearn library
#### 3.apply the column transforma() on the  data


## step 1

```python
num_features=['ApplicantIncome','CoapplicantIncome','LoanAmount']
cat_features=['Property_Area']
```

## step 2

```python
from sklearn.compose import ColumnTransformer
col_trans=ColumnTransformer(
                    [
                        ('numerical_pipeline',num_pipe,num_features),
                        ('categorial_pipeline',cat_pipe,cat_features)
                    ]
)
col_trans
```

## step 3

```python
data_col_trans=col_trans.fit_transform(dt_mar)
data_col_trans
```

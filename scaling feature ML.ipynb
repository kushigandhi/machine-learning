# method 2

```python
df['Age'].agg(['min','max'])
```

```python
df['Age_scale']=(df['Age']-df['Age'].min())/(df['Age'].max()-df['Age'].min())
```

```python
df
```

```python
df['Salary_scale']=(df['Salary']-df['Salary'].min())/(df['Salary'].max()-df['Salary'].min())
df
```

# 3 method

```python
df
```

```python
df['salary_scaled']=(df['Salary']-df['Salary'].mean())/(df['Salary'].std())
df
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
ss
```

# 4 method 

```python
df[['Age','Salary']].quantile([.25,.75])
```

```python
df['age_sc']=(df['Age']-df['Age'].quantile(.25))/(df['Age'].quantile(.75)-df['Age'].quantile(.25))
df
```

```python
df['sal_sc']=(df['Salary']-df['Salary'].quantile(.25))/(df['Salary'].quantile(.75)-df['Salary'].quantile(.25))
df
```

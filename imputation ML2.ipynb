# constant imputation

```python
df2=df.copy()
df2
```

```python
df2['Gender']=df2['Gender'].fillna('Female')
df2
```

```python
df2['Dependents']=df2['Dependents'].fillna('2')
df2
```

```python
df2['Self_Employed']=df2['Self_Employed'].fillna('No')
df2
```

```python
df2['LoanAmount']=df2['LoanAmount'].fillna('300.0')
df2
```

```python
df2['Married']=df2['Married'].fillna('yes')
df2
```

```python
df2['Loan_Amount_Term']=df2['Loan_Amount_Term'].fillna('Loan_Amount_Term')
df2
```

```python
df2['Credit_History']=df2['Credit_History'].fillna('Credit_History')
df2
                                                   
```

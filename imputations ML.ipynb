```python
# missing values treatment 
 #1.dropping missing value from the data
```

```python
data1=data.copy()
data1
```

```python
data=data.dropna()
data
```

```python
data.shape
```

```python
#drop value is last option in missing value 
#dropna() dropping the missing values deletes the complete rows in the data hence , dropna is last  
# option and its not a good strategy 
```

```python
# 2nd method-- constant imputation  
```

```python
data1['Salary']=data1['Salary'].fillna(93000.0)
data1
```

```python
data1['Age']=data1['Age'].fillna(40)
data1
```

```python
data1['Country']=data1['Country'].fillna('France')
data1
```

```python
# constant imputation can be used for both numerial and text values 
```

```python
#3rd method -- mean imputation
```

```python
data3=data.copy()
data3
```

```python
data3['Salary'].mean()
```

```python
data3['Salary']=data3['Salary'].fillna(data3['Salary'].mean())
data3
```

```python
data3['Age']=data3['Age'].fillna(data3['Age'].mean())
data3
```

```python
## 4 method--median imputation
```

```python
df4=data.copy()
df4
```

```python
df4['Salary'].median()
```

```python
df4['Salary']=df4['Salary'].fillna(df4['Salary'].median())
```

```python
df4
```

```python
df4['Age']=df4['Age'].fillna(df4['Age'].median())
df4
```

```python
# 5 method--mode imputation 
# mode give series output
```

```python
df5=data.copy()
df5
```

```python
df5['Country'].value_counts() #series output
```

```python
df5['Country'].mode()
```

```python
df5['Country']=df5['Country'].fillna(df5['Country'].mode()[0])
```

```python
df5
```

```python
df5['Age'].mode()
```

```python
df5['Age']=df5['Age'].fillna(df5['Age'].mode()[0])
df5
```

```python
#6 method--forward fill imputation 
# it works both on number and string values
```

```python
df6=data.copy()
df6
```

```python
df6=df6.fillna(method='ffill')
df6
```

```python
# 7 method-- backward fill imputation  
```

```python
df7=data.copy()
df7
```

```python
df7=df7.fillna(method='bfill')
df7
```

```python
#summary of missing value treatment 
## numericals variables all methods work on numerical values
## text variable mode,bfill,ffill works on text vales
## dropna deleting the missing value ( not recomended because entire row will details will be deleted)
 
    ### which is the best method
    ## for numerial variable  median imputation  is best 
    ## for text variable mode imputation is best 
```

```python
## mean median mode imputation  using sklearn lib
```

```python
df8=data.copy()
df8
```

```python
from sklearn.impute import SimpleImputer
```

```python
si=SimpleImputer(strategy='mean')
si
```

```python
df8['Age']=si.fit_transform(df8[["Age"]])
df8
```

```python
si=SimpleImputer(strategy='median')
si
```

```python
df8['Salary']=si.fit_transform(df8[["Salary"]])
df8
```

```python
si=SimpleImputer(strategy='most_frequent')
si
```

```python
df8['Country']=si.fit_transform(df8[["Country"]])
df8
```

```python
df8['Age']=si.fit_transform(df8[["Age"]])
df8
```

```python
df8['Salary']=si.fit_transform(df8[["Salary"]])
df8
```

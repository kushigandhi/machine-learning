```python
import pandas as pd
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
url=r'https://gist.githubusercontent.com/ryanorsinger/0c702f67dfe59346cd4677cd24f7aec7/raw/429f0d47856670e57b31d1f78a1abc54c889c7ca/tips.csv'
url
```

```python
data=pd.read_csv(url)   #sns.loadcsv
data.head()
```

```python
data.isnull().sum()
```

```python
from sklearn.preprocessing import RobustScaler
rs=RobustScaler()
rs
```

```python
data[['total_bill','tip']]=rs.fit_transform(data[['total_bill','tip']])
data
```

```python
data=pd.get_dummies(data,dtype='int',drop_first=True)
data
```

```python
x=data.drop(columns='tip')
x
```

```python
y=data['tip']
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsRegressor
knm= KNeighborsRegressor(n_neighbors=9)
knm
```

```python
knm.fit(x_train,y_train)
```

```python
y_pred=knm.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score
R2=r2_score(y_test,y_pred)
R2
```

```python
from sklearn.metrics import root_mean_squared_error
root_mean_squared_error(y_test,y_pred)
```
